```python
# plt.style.use("ggplot")
plt.rcParams["font.family"] = ["Microsoft JhengHei"] # 正黑體
# plt.rcParams["font.family"] = ["Arial"] # 正黑體
plt.rcParams["axes.unicode_minus"] = False           # 可以顯示x, y 軸負數

X = np.linspace(-np.pi, np.pi, 25, endpoint=True)
S, C, T = np.sin(X), np.cos(X), np.tan(X)
#均勻地切出 25 個點。這就是你的 X 軸。把這 25 個點，分別帶入 $\sin(x)$、$\cos(x)$、$\tan(x)$ 計算出對應的高度（Y 軸）。

plt.plot(X, S,'*')
plt.plot(X, C)
# plt.plot(X, T)

plt.title('this is title 這是標題')
plt.xlabel('x label(徑度)'), plt.ylabel('y label(值)')
plt.grid(alpha=.2)          # try plt.grid(axis='x')

plt.savefig(r'.\img\grid.jpg', dpi=150)  # 要寫在 plot show 之前
plt.savefig(r'.\img\grid.pdf', dpi=150, format='pdf')  # 要寫在 plot show 之前 (format用來二次檢查，不寫也可以)
plt.show()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use("ggplot")

plt.figure(figsize=(8, 5))
# plt.subplot(n_rows, n_cols, plot_num)
# 建立一張寬 8、高 5 的大畫布（Figure）
plt.subplot(2, 2, 1) # 啟用 2x2 網格中的第 1 格（左上角）
plt.plot([0, 1], [0, 1]) # 在這格裡面畫線
plt.title('plt221')

plt.subplot(222)# 這裡寫 222（沒逗號）效果完全一樣
plt.plot([1, 1], [0, 2])
plt.xlabel('222 x label')

plt.subplot(223)
plt.plot([2, 1], [3, 3])
plt.title('plt223')

plt.subplot(224)
plt.plot([4, 1], [1, 4])
plt.title('title224')


#tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area.
plt.tight_layout()
# tight_layout 會自動調整子圖參數，讓子圖能完美塞進畫布區域，避免文字重疊。
```

## Text & Annotation(標註)

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# 繪製一個餘弦曲線
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

# 繪製一個黑色，兩端縮排的箭頭
ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            xycoords='data', arrowprops=dict(facecolor='r', shrink=0.05))
ax.text(1.0, 1.25, "Peak Value",fontsize=12, c='r')

ax.set_ylim(-2, 2)
plt.show()
```

## Bar side by side一個靠著一個

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
N = 5

men_means = (20, 35, 30, 35, 27)
women_means = (25, 32, 34, 20, 25)

ind = np.arange(N) 
width = 0.45      
M=plt.bar(ind, men_means, width, label='Men_means')
plt.bar_label(M, padding=1, color='b')

W=plt.bar(ind + width, women_means, width, label='Women_means')
plt.bar_label(W, padding=1, color='r')

plt.ylabel('Scores')
plt.title('Scores by group and gender')

plt.xticks(ind + width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend(loc='best')
plt.show()
```

## Bar stack

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

countries = ['USA', 'GB', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])
ind = list(range(len(countries)))

print(ind)

g=plt.bar(ind, golds, width=0.8, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar_label(g, padding=1, color='g')

s=plt.bar(ind, silvers, width=0.8, label='silvers', color='silver', bottom=bronzes)
plt.bar_label(s, padding=1, color='r')

b=plt.bar(ind, bronzes, width=0.8, label='bronzes', color='#CD853F')
plt.bar_label(b, padding=1, color='b')

plt.xticks(ind, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right")
plt.title("2012 Olympics Top Scorers")

plt.show()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
 
# Create data
N = 60
g1 = (0.6+0.6*np.random.rand(N), np.random.rand(N))
g2 = (0.4+0.3*np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.0+0.3*np.random.rand(N), 0.3*np.random.rand(N))

data = [g1, g2, g3]
colors = ["red", "green", "blue"]
groups = ["coffee", "tea", "water"]

for data, color, group in zip(data, colors, groups):
    x, y = data
    plt.scatter(x, y, alpha=0.9, c=color, edgecolors='none', s=30, label=group)

plt.legend(loc=2)
plt.title('Matplot scatter plot')
plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# countries = ["china", "usa", "france", "russia",
# df = pd.read_csv("./datasets/titanic/train.csv")
df = pd.read_csv("./data/titanic.csv")
# print(f'{df.info()}\n\n'
#       f'df.head() :\n{df.head()}')
# print(df)

df=df.corr(method='pearson', numeric_only=True)
print(df)

fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(df, interpolation='nearest')
fig.colorbar(im, orientation='vertical', fraction = 0.05)

ax.set_xticks(np.arange(len(df.columns)))
ax.set_xticklabels(df.columns)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
ax.set_yticks(np.arange(len(df.index)))
ax.set_yticklabels(df.index)
for i in range(len(df.columns)):
    for j in range(len(df.index)):
        ax.text(j, i, df.iloc[i, j].round(2), ha="center", va="center", 
                color="w", fontsize=18)
ax.set_title("Corr")
plt.show()
```

# 補充教材 : 3D, Pandas vs Matplotlib

## 函式使用介紹：
plot_surface(X, Y, Z, *args, **kwargs)
> * X, Y, Z ：分別是三個代表 X, Y, Z 座標的 2D array
> * cstride, rstride ：分別對應 column 與 row 方向的降採樣步伐，也就是繪圖的時候多少格要畫在同個網格內
> * cmap ：Colormap，輸入一個字串對應一個3D surface的顏色

colorbar(mappable, **kwargs)
> * mappable ：colorbar 要加入的圖
> * shrink ：控制 colorbar 的高度(預設為 1.0)
> * aspect ：控制 colorbar 的寬高比(預設為 20)

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

# 建立 3D 圖形
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

# 產生測試資料
X, Y, Z = axes3d.get_test_data(.05)

# 繪製 Wireframe 圖形
ax1.plot_wireframe(X, Y, Z, rstride=10, cstride=10, lw=.5)
ax2.plot_surface(X, Y, Z, cmap='seismic')

# 顯示圖形
plt.show()
```

```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
# ax = Axes3D(fig)
# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
# height value
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))

# I think this is different from plt12_contours
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2, 2)

plt.show()
```
