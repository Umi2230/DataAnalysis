# Mod 06 Introducing `Series` Objects
>* Pandas 在異質數據的讀取、轉換和處理上，都讓分析人員更容易處理，例如：從`列欄試算表`中找到想要的值。
>* Pandas 提供兩種主要的資料結構，`Series 與 DataFrame`。Series 顧名思義就是用來處理序列相關的資料，主要為建立`索引的一維陣列`。DataFrame (excel表)則是用來處理結構化(Table like)的資料，有`列 row 索引與欄 column 標籤`的二維資料集，例如關聯式資料庫、CSV 等等。(如果是datafrme ，即使看到的資料只有一行，一樣是二維資料)
>* 透過載入至 Pandas 的資料結構物件後，可以透過結構化物件所提供的方法，來快速地進行`資料的前處理`，如資料補值，空值去除或取代......等。
>* 更多的輸入來源及輸出整合性，例如：可以從`資料庫讀取資料進入` Dataframe，也可將處理完的資料`存回資料庫`。

## 6-1 The Pandas Series Object
> ### list vs. array vs. Series : `array like`
> >><img src="./img/series01.png"  style='width:80%'>

```python
# float
lst1 = [0.25, 0.5, -0.75, 1]      # list
np_array = np.array(lst1)         # array,    try reshape

ser1 = pd.Series(lst1)            # list to series
ser2 = pd.Series(np_array)        # array to series

print(f'lst1\t\t= {lst1}\n'
      f'np_array\t= {np_array}\n\n'
      f'ser1 =\n{ser1}\n\n'
      f'type(ser1) : {type(ser1)}\n\n'
      f'ser2 =\n{ser2}')
```

## 6-2 boolean / logical & arithmetic operation

```python
print(f'ser1 =\n{ser1}\n\n'
      f'ser1[ser1>0] =\n{ser1[ser1>0]}\n\n'
      f'ser1*2 =\n{ser1*2}\n\n'
      f'np.exp(ser1) =\n{np.exp(ser1)}')
```

## 6-3 成員運算子(Membership Operators)

```python
print(f'{ser1}\n\n'
      f'2 in ser1\t: {2 in ser1}\n'     # 2 is index, not value
      f'0.5 in ser1\t: {0.5 in ser1}')  # 1, 2 are indexes, not values
```

### dtype : object

```python
ser1 = pd.Series([0.25, 0.5, 'hello', 1.0])  # dtype : object
ser1, print(ser1)
```

## 6-7 loc 是 key, iloc 是 intgrate-location(整數位置)
>### loc - `label` indexing 標籤索引 (標籤用)
>### iloc - `positional` indexing 位置索引 (一定是 int 整數用)
>><img src="./img/loc_iloc.jpg"  style='width:100%'>

### loc

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])

print(f'{data}\n\n'
      f'{data.loc[3]}\n\n' 
      f'{data[1:3]}\n\n'   # position index 沒寫就是iloc -->不含3(不含終點)
      f'{data.loc[1:3]}') # 包含 3(終點會跑出來)
```

### iloc

```python
print(f'{data}\n\n'
      f'{data.iloc[3]}\n\n'
      f'{data[1:3]}\n\n'
      f'{data.iloc[1:3]}') # 不包含 3
```

### index=['5','4','3','2']

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=['5','4','3','2'])
print(f"{data}\n\n"
      f"{data.loc['3']}\n\n"
      f"{data['3':'5':-1]}\n\n"
      f"{data.loc['5':'3']}") # include 3
```

### iloc

```python
print(f'{data}\n\n'
      f'{data.iloc[3]}\n\n'
      f'{data.iloc[1:3]}') # 不包含 3
```

## 7-3 random sample : rows or [columns]
> DataFrame.sample(n=None, frac=None, replace=False, weights=None, random_state=None, axis=None)

```python
# s1 = pd.Series(range(10, 15), ind_char)
s1 = pd.DataFrame(np.arange(50).reshape(-1, 5),  
                   columns=['Ohio', 'Texas', 'California', 'illinos', 'New York'])
print(s1, '\n')
s1['New York']=s1['New York'].apply(lambda x:1 if x<25 else 2)
s1
```

### replace = True : 抽樣完放回

```python
print(f'frac=.6 :\n{s1.sample(frac=.6)}\n\n'   # replace =1, replace = False 抽樣60%
      f'n=5, axis=0, replace=True, random_state=2 :\n{s1.sample(n=5, axis=0, replace=True, random_state=2)}\n\n'
      #抽樣5筆 0軸 可重複 亂數固定
      f'frac=.6, axis=1 :\n{s1.sample(frac=.6, axis=1)}\n\n'
      # 僅水平抽60%
      f'frac=2, replace=True :\n{s1.sample(frac=2, replace=True)}')   # frac =2, replace=True  
      #抽200% 可重複
```

### build train and test sets

```python
train = s1.sample(frac=.7) #抽七成
test = s1.drop(train.index) #然後把他刪掉
print(f's1 :\n{s1}\n\n'
      f'train :\n{train}\n\n'
      f'test :\n{test}')
```

## 11-1 df aggregation
> 對於 DataFrame，`預設`情況下，聚合函數傳回每個 `axis=0` 欄位的結果。

```python
rng = np.random.RandomState(42)
df = pd.DataFrame({'A': rng.rand(5), 'B': rng.rand(5)})

print(f'df =\n{df}\n\n'
      f'df.count() =\n{df.count()}\n\n'
      f'df.mean() =\n{df.mean()}\n\n'
      f'df.median() =\n{df.median()}\n\n'
      f'df.min() =\n{df.min()}\n\n'
      f'df.max() =\n{df.max()}\n\n'
      f'df.std() =\n{df.std()}\n\n'
      f'df.quantile(.5) =\n{df.quantile(.5)}')
```

### agg function vs. describe (聚合)
> agg 返回的也是一個DataFrame。很多功能用 sum、mean 等等也可以實現。但是 agg 更加簡潔, 而且傳給它的函數可以是字串，也可以自訂，參數是column對應的子DataFrame。

```python
print(f"df.agg('mean') :\n{df.agg('mean')}\n{'-'*60}\n"
      f"df.agg(['count', 'mean', 'std', 'min', 'max'] :\n{df.agg(['count', 'mean', 'std', 'min', 'max'])}\n{'-'*60}\n"
      f'df.describe() :\n{df.describe()}')
```

```python
print(f'df :\n{df}\n{"-"*60}\n'
      f"df.mean(axis='columns') :\n{df.mean(axis='columns')}\n{'-'*60}\n"
      f'df.mean(axis=1) :\n{df.mean(axis=1)}')
```

## 10-3 GroupBy Mechanics
> 用 groupby 進行資料分組，可以看做是基於行（或者說是index）操作的話，則 agg 函數則是基於`列`的聚合操作。

> groupby 返回的是一個 DataFrameGroupBy 結構，這個結構必須調用彙總函式（如sum）之後，才會得到結構為 Series 的資料結果。用 groupby 函數後，要接類似 sum、mean 等彙總函式才能輸出

><img src="./img/groupby.jpg"  style='width:100%'>

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
grouped = df['data1'].groupby(df['key1'])
grouped  # object 物件
```

```python
grouped.mean(), grouped.sum()
```

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
df, means
```
