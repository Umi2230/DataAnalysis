# Mod. 01 NumPy基礎 ：陣列和向量化計算

> NumPy (Numerical Python) 中, ndarray (numpy data array) 在儲存數據的時候，數據與數據的地址都是`連續的`，這樣就給使得批次運算元組元素時`速度更快`。

> * ndarray 在儲存元素時記憶體可以連續，而 python 原生 list 就只能通過定址方式找到下一個元素，這雖然也導致了在通用效能方面 Numpy 的 ndarray 不及 Python 原生 list，但在科學計算中，Numpy 的 ndarray 就可以`省掉很多回圈語句`，程式碼使用方面比 Python 原生 list 簡單的多。

> * ndarray `支援並行化運算（向量化運算）`numpy 內建了並行運算功能，當系統有多個核心時，做某種計算時，numpy 會自動做`並行計算`

> * 效率遠高於純 Python 程式碼, Numpy 底層使用 C 語言編寫，內部解除了GIL（全域性直譯器鎖），其對陣列的操作速度不受 Python 直譯器的限制，所以，`其效率遠高於純 Python 程式碼`。

> * 方便有用的線性代數 (Linear Algebra) 及傅立葉轉換 (Fourier Transform) 能力, 利用NumPy Array 替代 Python List 可定義任意的數據型態 (Data Type)，使得能輕易及無縫的與多種資料庫整合

>><img src="./img/processingSpeed.jpg">

> 在邁向資料科學 (Data Science)、機器學習 (Machine Learning)、深度學習 (Deep Learning) 及人工智慧 (Artificial Intelligence) 的路上， Python 還有許許多多有趣好玩的套件 (例如：Pandas、SciPy、Sympy、Matplotlib、Scikit-learn 等），這些與資料科學相關的重要套件幾乎都是架構在 Numpy 基礎上做應用，因此學會 NumPy 對後續的學習及理解將會有很大的幫助。

## speed : list vs. array 

```python
import numpy as np
import sys
my_arr = np.arange(1000000)
my_list = list(range(1000000))
# for i in my_list[:10]:
#     print(i)

#%time計算這一行跑的時間
%time for _ in range(10): my_arr2 = my_arr * 2 #執行效率較高
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]

print(f'\narray\t: {sys.getsizeof(my_arr):,} bytes\n'
      f'list\t: {sys.getsizeof(my_list):,} bytes')
```

```python
''' list '''
%timeit ar2=np.array([[0,3,5],[2,8,7]]) # 2D array list 

''' tuple ''' #函數計算時速度較快(特性:不可變 記憶體較小)
%timeit ar2=np.array(((0,3,5),(2,8,7))) # 2D array tuple
# ar2=np.array([[0,3,5],[2,8,7]])
```

### Inspecting Your Array : `array like`

```python
data1 = [6, 7.5, 8, 0, 1]  # list
arr1 = np.array(data1)     # list to array

data2 = [[1, 2, 3, 4], 
         [5, 6, 7, 8]]  # 2 index list
arr2 = np.array(data2)              # 2d array

print(f'arr1\t\t= {arr1}\n'
      f'arr1.ndim\t= {arr1.ndim}\n' #一維
      f'arr1.shape\t= {arr1.shape}\n' #5列1行
      f'arr1.size\t= {arr1.size}\n' #五筆資料
      f'arr1.dtype\t= {arr1.dtype}\n' #在電腦記憶體中所佔用的位元數（Bits），64bit
      f'arr1.itemsize\t= {arr1.itemsize} bytes\n') #64bit = 8byte

print(f'arr2 =\n{arr2}\n\n' 
      f'arr2.ndim\t= {arr2.ndim}\n'
      f'arr2.shape\t= {arr2.shape}\n'
      f'arr2.size\t= {arr2.size}\n'
      f'arr2.dtype\t= {arr2.dtype}\n'
      f'arr2.itemsize\t= {arr2.itemsize} bytes\n')   
# float64類型元素的itemsize值為 8=(64/8)，complex32類型元素的itemsize值為 4=(32/8)
```

# Mod 02 指標和切片 Indexing & Slicing
> Numpy 中多維陣列的切片操作與 Python 中 `list 的切片操作一樣`，同樣由 `start, stop, step` 三個部分組成

## 2-1 Basic Indexing & Slicing

> ### Illustrate the scope of indexing in NumPy
> <img src="./img/index.jpg" style='width:100%'></img>

> #### `index : arr[row, column, ....]`
> #### `fancy index : arr[[3,4,1], [1,2,3], ....]`
> #### `slice : arr[row:row:step, column:column:step, ....]`
> #### 以上三種方式`可混用`
> <img src="./img/np_arrays.png" style='width:100%'></img>

> <img src="./img/index_2d.jpg" style='width:100%'></img>

```python
import numpy as np
arr = np.arange(10)
print(f'arr\t\t= {arr}\n'
      f'arr[5]\t\t= {arr[5]}\n'
      f'arr[5:8]\t= {arr[5:8]}\n'
      f'arr[5:8:2] \t= {arr[5:8:2]}')

arr[5:8] = 12
print(f'arr\t\t= {arr}')
```

### Copy & Deep Copy
> ### slice point back to original array

```python
arr_slice = arr[5:8]
arr, arr_slice
```

```python
# point to original arr_slice.  leverage back to origianl
arr_slice[1] = 12345
arr, arr_slice   
#原始的值也會改變
```

```python
arr_slice[:] = 64
arr, arr_slice 
```

```python
import numpy as np

a = np.arange(4)
b = c = a; d = b

print(f'a = {a}\n'         
      f'b = {b}\n'         
      f'c = {c}\n'         
      f'd = {d}\n')       

d[1:3] = [22, 33]   # array([0, 22, 33,  3])
print(f'a = {a}\n'        
      f'b = {b}\n'        
      f'c = {c}\n'        
      f'd = {d}')      
```

> 利用 numpy 的聚合函數 (Aggregate Functions) 於統計領域是很方便的，常見聚合函數有最大值 (max)、最小值 (min)、中位數 ( median)、平均數 (mean)、標準差 (std)、在 numpy 的使用上是很常見的。

Functions	| Description
----------:|--
np.mean()  | 計算指定軸上的算術平均值
np.std()   | 計算指定軸上的標準差
np.var()   | 計算指定軸上的變異數
np.sum()   | 計算給定軸上數組元素的和
np.prod()  | 傳回給定軸上數組元素的乘積
np.cumsum()  | 傳回給定軸上元素的累積和
np.cumprod() | 傳回給定軸上元素的累積乘積
np.min(), np.max()| 傳回陣列的最小值/最大值，或傳回某個軸上的最小值
np.argmin(), np.argmax()| 傳回某個軸上最小值/最大值的索引
np.all()  | 測試給定軸上所有陣列元素是否都為 True
np.any()  |測試沿給定軸的任何數組元素是否計算為 True

## 5-1 常態分佈 normal distribution
> 常態分佈是一種概率分佈。常態分佈是具有兩個參數 μ 和 σ 的連續型隨機變數的分佈，第一參數 μ 是常態分佈的隨機變數的均值，第二個參數 σ 是此隨機變數的標準差(方差)，所以常態分佈記作 N(μ，σ )。μ 決定了其`位置`，其標準差 σ 決定了分佈的`幅度`。`當 μ = 0, σ = 1 時的常態分佈是標準常態分佈`。

> σ 爲標準差 : 是在概率論和統計方差衡量一組數據時 `離散程度` 的度量

>> <img src="./img/normalDistribution.png" style='width:100%'></img>
>> <img src="./img/normalDistribution01.png" style='width:100%'></img>
>> <img src="./img/norm_dist.png" style='width:100%'></img>

### 常態分佈 random.randn
> rand函數根據`給定維度`產生[0,1)之間的數據，包含0，不包含1<br>
#### randn(d0, d1, ..., dn)    d0, d1... 維度   mean=0, stdev=1

### random.uniform() : floating & int numbers
> 從一個均勻分佈 low, high) 中隨機採樣，注意定義域是左閉右開，即包含low，不含high.
>> * low: 取樣下界，float類型，預設值為0<br>
>> * high: 取樣上界，float類型，預設值為1<br>
>> * size: 輸出樣本數目，為int或元組(tuple)型，例如，size=(m,n,k), 則輸出$m\times n\times k$個樣本，缺省時輸出1個值。
 > * 傳回值：ndarray類型，其形狀和參數size中描述一致。

```python
x1 = np.random.uniform(low=10.0, high=15.0, size=1000000)
# x1 = np.random.uniform(0.0, 1.0, 10)  # same

# [low ~ high) integer
x2 = np.random.randint(low=0, high=5, size=1000000)
print(x1[:10], '\n', x2[:10])

# 畫圖看分佈狀況
plt.figure(figsize=(10, 3))  # 建立畫布
plt.subplot(1, 2, 1)
plt.hist(x1, 100)           # 繪製直方圖

plt.subplot(1, 2, 2)
plt.hist(x2, 100)           # 繪製直方圖

plt.show()                  # 顯示影像
```

### 距離 distance / 相似度 similarity
> ### 距離的定義

>><img src="./img/distance.jpg"  style='width:80%'></img>

$$if\quad A = [1,2,0,4],\quad B = [3,2,1,0]$$
#### 1. Euclidean distance(歐基里德距離)
$$d(i, j) = \sqrt{(x_{i1} - x_{j1})^2 + (x_{i2} - x_{j2})^2 + \dots + (x_{in} - x_{jn})^2}$$

$$d = \sqrt{(1-3)^2+(2-2)^2+(0-1)^2+(4-0)^2}$$

```python
import numpy as np
# two points
# a = np.array((1,0));   b = np.array((0,1))
a = np.array((2, 3, 6, 9, 5));   b = np.array((5, 7, 1, 2, 2))

# distance b/w a and b
np.linalg.norm(a-b)
```

#### 2. Cosine Similarity : 衡量 cosine($\theta$) 的大小

$$if\quad A = [1,2,0,4],\quad B = [3,2,1,0]$$
$$\cos{\theta} = \frac{A \cdot B}{|A||B|}$$<br>
$$\cos{\theta} = \frac{1 \cdot 3 + 2 \cdot 2 + 0 \cdot 1 + 4 \cdot 0} {\sqrt{1^2+2^2+0^2+4^2} \cdot \sqrt{3^2+2^2+1^2+0^2}}$$

>><img src="./img/vector.png"  style='width:70%'></img>

> * Square difference 平方差, 這是最常見的數學公式

> * Correlation 相關性 : 計算 dot product (內積)，可以想成是計算兩個向量在空間中的距離有多近，就是用 cosine 去算夾角，cosine `值越大代表夾角越小`，代表越接近 ( 假設模板影像是 10*10 的影像，可以被看作是 100 維的向量，每一維是像素的值 )<br>
$$cos \theta = \frac{a \cdot b}{|a||b|}\quad \Longrightarrow \quad \theta = cos^{-1}\left({\frac{a \cdot b}{|a||b|}}\right)$$

$$if \quad a=a_1i+a_2j+a_3k \quad and \quad b=b_1i+b_2j+b_3k $$
$$ then\quad \theta = cos^{-1}\left(\frac{a_1b_1+a_2b_2+a_3b_3}{\sqrt{a_1^2+a_2^2+a_3^2}\sqrt{b_1^2+b_2^2+b_3^2}}\right) $$

>> a dot b 越大, $cos \theta$ 夾角越小

>><img src="./img/cosine.png"  style='width:90%'></img>

>> Correlation coefficient 與 Correlation 差別只在於計算內積時還要減去各個向量的平均值，如此一來相關性就會被放大

```python
import numpy as np
import matplotlib.pyplot as plt

def cosine_sim(a, b) :
    num = np.dot(a, b)
    denom = np.linalg.norm(a) * np.linalg.norm(b)
    if denom == 0:
        return np.degrees(np.arccos(0.0))   # 或改成 np.nan
    return np.degrees(np.arccos(num / denom))
    
v1 = np.array([1, 2, 3]);    v2 = np.array([4, 5, 6])
print(f"θ = {cosine_sim(v1, v2):.2f}°")  # -> 0.9746...

# 3D 圖設定
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x');    ax.set_ylabel('y');    ax.set_zlabel('z')

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='blue', label='v1', arrow_length_ratio=0.08)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='red', label='v2', arrow_length_ratio=0.05)
ax.text(0.5, 0.5, 0.5, f"θ = {cosine_sim(v1, v2):.2f}°", fontsize=12)
ax.text(v1[0], v1[1], v1[2], f"v1 {tuple(v1)}", fontsize=12)
ax.text(v2[0], v2[1], v2[2], f"v2 {tuple(v2)}", fontsize=12)
plt.title("3D Vector Angle Visualization");    ax.legend()

ax.set_xlim([0, max(v1[0], v2[0]) + 1])
ax.set_ylim([0, max(v1[1], v2[1]) + 1])
ax.set_zlim([0, max(v1[2], v2[2]) + 1])
```

### hamming
漢明距離是用來衡量 兩個等長序列（通常是二進位序列）之間不同位元的數量
>><img src="./img/hamming.png"  style='width:70%'></img>

```python
import numpy as np

def hamming_distance(a, b):
    a = np.asarray(a);    b = np.asarray(b)
    print(a!=b, np.bitwise_xor(a, b), np.where(a!=b,1,0))
    return np.sum(a!=b)

print(hamming_distance([1,0,1,1], 
                       [1,1,0,1]))  # 2
```
