## 1-12 for loop
>### `迭代` :　是程式中對一組指令（或一定步驟）的重複。它既可以被用作通用的術語（與「重複」同義）

```python
for i in [6, 5, 15, 3, 2, 10]:
    print(i)
```

```python
for i in range(2, 100, 10) :
    print (i)
```

```python
for i in range(1, 10, 2):
    print(f"{'a'*i:^20s}")
```

## 1-13 zip

```python
names = ["A", "B", "C"]# 第一個 List
values = [11, 23, 46]# 第二個 List

# 使用 zip 同時迭代兩個 List
for x, y in zip(names, values):
    print(x, y)
for z in zip(names, values):
    print(z)
```

## 1-14 map

```python
lst=[2, 3, 2, 6, 3, 1, 3]

m_1=map(lambda x:3*x+2, lst)
print(list(m_1))

m_2=map(lambda x:float(x), lst)
print(list(m_2))
```

```python
w=['Mary', 'John', 'Tom', 'Rose']
m_3=map(lambda x:x.upper(), w)
print(list(m_3))
```

## 1-15 filter

```python
lst=[2, 3, 2, 6, 3, 1, 3]
m_2=filter(lambda x:x>=3, lst)
print(list(m_2))
```
